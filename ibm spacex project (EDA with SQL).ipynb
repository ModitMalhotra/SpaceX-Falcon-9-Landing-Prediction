```python
import pandas as pd

df = pd.read_csv(
    "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv"
)
df.to_sql("SPACEXTBL", con, if_exists="replace", index=False, method="multi")
```

```python
%sql DROP TABLE IF EXISTS SPACEXTABLE;
```

```python
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null
```

```python
%sql SELECT DISTINCT launch_site from SPACEXTABLE
```

```python
%sql SELECT * FROM SPACEXTABLE WHERE launch_site LIKE 'CCA%' LIMIT 5
```

```python
%sql SELECT SUM(PAYLOAD_MASS__KG_) FROM SPACEXTABLE WHERE Customer LIKE '%NASA (CRS)%'
```

```python
%sql SELECT AVG(PAYLOAD_MASS__KG_) FROM SPACEXTABLE WHERE Booster_Version LIKE '%F9 v1.1%'
```

```python
%sql SELECT MIN(Date) from SPACEXTABLE WHERE Landing_Outcome LIKE '%success%' AND Landing_Outcome LIKE '%ground pad%'
```

```python
%sql SELECT Booster_Version from SPACEXTABLE WHERE Landing_Outcome LIKE '%drone ship%' AND Landing_Outcome LIKE '%success%' AND PAYLOAD_MASS__KG_ >4000 AND PAYLOAD_MASS__KG_ < 6000
```

```python
%sql SELECT Mission_Outcome, COUNT(Mission_Outcome) FROM SPACEXTABLE GROUP BY Mission_Outcome
```

```python
%sql SELECT Booster_Version FROM SPACEXTABLE WHERE PAYLOAD_MASS__KG_=(SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTABLE)
```

```python
%sql SELECT substr(Date,6,2) as Month, Landing_Outcome, Booster_Version, Launch_Site FROM SPACEXTABLE WHERE Landing_Outcome LIKE '%drone ship%' AND Mission_Outcome = 'Failure' AND substr(Date,0,5) = '2015'
```

```python
%sql SELECT Landing_Outcome, COUNT(Landing_Outcome) as Count FROM SPACEXTABLE WHERE Date BETWEEN '2010-06-04' AND '2017-03-20' GROUP BY Landing_Outcome ORDER BY Count DESC
```

```python
%sql SELECT COUNT(*) FROM SPACEXTABLE WHERE Launch_Site = 'CCAFS SLC-40'
```

```python
%sql SELECT COUNT(*) FROM SPACEXTABLE WHERE Orbit = 'GEO'
```

```python
%sql SELECT COUNT(*) FROM SPACEXTABLE WHERE Landing_Outcome LIKE '%drone ship%' AND Mission_Outcome = 'Success'
```
